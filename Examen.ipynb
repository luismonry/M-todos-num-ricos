Calcule la temperatura a la cual una solución al 60% tiene una densidad de 1.48XY Donde X y Y son los dos últimos dígitos de su número de cuenta. 318259404 por lo tanto una densidad de **1.4804**
Use la interpolación del Spline Cúbico.

```python
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
x=np.array ([10,30,60,100])
y=np.array ([5,20,40,70])
z=np.array([[1.0344,1.0281,1.0140,0.9888],\
            [1.1453,1.1335,1.1153,1.0885],\
            [1.3103,1.2953,1.2732,1.2446],\
            [1.6923,1.6014,1.5753,1.5417]])
```

```python
import numpy as no
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
f= interpolate.interp2d(x,y,z,kind='cubic')#Vector densidades a 60% de concentración
xi=([10,30,60,100])
yi=60
zi=f(xi,yi)
print(zi)
```
