La concentración de un reactivo $A$ varía con el tiempo según la siguiente ecuación

$$
\frac{dC_A}{dt}=-9420e^{-\frac{3500}{298}}C_A
$$

que no depende de la temperatura. Se realiza un experimento y al cabo de 25 min se encuentra en el sistema 3.4 mol/l de $A$ ¿cuál era la concentración inicial? Grafique la variación de la concentración con el tiempo.

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(C,t):
  k1 =9420*np.exp(-3500/298)
  dcA =-k1*C[0]
  return [dcA]
```

```python
x0 =0 # valor inicial de tiempo

x1 =25 # valor final del tiempo
y1 =[3.4] # condiciones iniciales de
tiempo=np.linspace(x1,x0)
# llamada a la funcion odeint
sol = odeint(f,y1,tiempo )
sol
```

```python
pmax,idx = np.max(sol[:,0]),np.argmax(sol[:,0])
t=tiempo[idx]
print('tiempo =',t,'min\nproducto max = ',pmax,'mol/L')
```

```python
fig,ax=plt.subplots(1,figsize=(12,6))
ax.minorticks_on()
plt.plot(tiempo,sol[:,0],label = '$Reactivo A$')
plt.legend(fancybox=True,shadow=True,loc='upper center')
plt.grid(which='major',linestyle='-.',linewidth=0.5,color='red')
plt.grid(which='minor',linestyle='--',linewidth=0.5,color='gray')
plt.title('Variación en la concentración del Reactivo A',fontsize=18)
plt.xlabel('tiempo [min]')
plt.ylabel('concentración [mol/L]')

plt.show()
```

La concentración de un reactivo $B$ varía con el tiempo según la siguiente ecuación 

$$
\frac{dC_B}{dt}=-5\cdot 10^{15}e^{-\frac{15000}{T}}C_B^2
$$

y la temperatura varía con el tiempo según la ecuación $T=298+5t$. En el sistema se inicia con una concentración de $B=7mol/l$ ¿cuál es la concentración al cabo de 25 min? Grafique la variación de $C_B$ - $t$

```python
5
```

```python
def f(C,t):
  k2 =-5*10**15*np.exp(-15000/(298+5*t))
  dcB =k2*C[0]**2
  return [dcB]
```

```python
x0=0 # valor inicial de tiempo
#y0 =[ B ]
y0 =7 # condiciones iniciales de b
x1 =25 # valor final del tiempo
tiempo =np.linspace (x0,x1)
# llamada a la funcion odeint
sol = odeint(f,y0,tiempo)
sol
```

```python
fig,ax=plt.subplots(1,figsize=(12,6))
ax.minorticks_on()
plt.plot(tiempo,sol[:,0],label = '$Reactivo B$')
plt.legend(fancybox=True,shadow=True,loc='upper right')
plt.grid(which='major',linestyle='-.',linewidth=0.5,color='red')
plt.grid(which='minor',linestyle='--',linewidth=0.5,color='gray')
plt.title('Variación en la concentración del Reactivo B',fontsize=18)
plt.xlabel('tiempo [min]')
plt.ylabel('concentración [mol/L]')

plt.show()
```
